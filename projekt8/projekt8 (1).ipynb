```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
sp.init_printing()
```

```python
S = sp.Symbol('S',real=True) #narażeni
I = sp.Symbol('I',real=True) #chorzy
R = sp.Symbol('R',real=True) #usunięci
beta = sp.Symbol('beta',real=True) #zarazliwosc
gamma = sp.Symbol('gamma',real=True) #zdrowotnosć
t=sp.Symbol('t',real=True)

#eq = sp.Eq(S(t).diff(t), -beta*S(t)*I(t))
#eq1 = sp.Eq(I(t).diff(t), beta*S(t)*I(t) - gamma*I(t))
#eq2 = sp.Eq(R(t).diff(t), gamma*I(t))
#eq
#eq1
#eq2
```

```python
macierz = sp.lambdify((t,[S,I,R],beta,gamma),[-beta*S*I,beta*S*I-gamma*I,gamma*I])
ts = np.linspace(0, 50, 5000)
```

```python
S0 = 0.999
R0= 0
I0 = 0.001
B = 2
G = 1.5
solution = solve_ivp(macierz, (0, 50), [S0,I0,R0], t_eval=ts, args=[B, G])
plt.plot(ts, solution.y.T)
plt.grid()
```
