```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
sp.init_printing()
```

```python
x = sp.Function('x',real=True)
t=sp.Symbol('t',real=True)
w0=sp.Symbol('omega_0',real=True,positive=True)



Beta = sp.Symbol('beta',real=True, positive=False) #parametr tłumienia
A=sp.Symbol('A',real=True)
w1=sp.Symbol('omega_1',real=True,positive=True) #czestotliwosc wymuszenia

xp=sp.Symbol('x_p',real=True)
vp=sp.Symbol('v_p',real=True)
#both
eq = sp.Eq(x(t).diff(t, t)+ 2*Beta*x(t).diff(t) + w0**2 * x(t) , A*sp.cos(w1*t))
eq
#z wymuszeniem
eq1 = sp.Eq(x(t).diff(t, t) + w0**2 * x(t) , A*sp.cos(w1*t))
eq1
#z tłumieniem
eq2 = sp.Eq(x(t).diff(t, t)+ 2*Beta*x(t).diff(t) + w0**2 * x(t) , 0)
eq2
```

```python
#z wymuszeniem
sol1=sp.dsolve(eq1,ics = {x(0): xp, x(t).diff(t).subs(t, 0): vp})
sol1.simplify()
```

```python
vis = sp.lambdify(t, sol1.rhs.subs({xp: 1, vp: 1, w0: 1,w1: 2,A: 5}))
ts = np.linspace(0, 50, 500)
plt.plot(ts, vis (ts))
plt.grid()
```

```python
vis  = sp.lambdify(t, sol1.rhs.subs({xp: 1, vp: 1, w0: 1,w1: 2,A: 10}))
ts = np.linspace(0, 50, 500)
plt.plot(ts, vis (ts))
plt.grid()
```

```python

sol2=sp.dsolve(eq2,ics = {x(0): xp, x(t).diff(t).subs(t, 0): vp})
sol2=sol2.simplify()
sol2
```
